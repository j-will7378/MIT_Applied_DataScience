```python
from ipython_genutils.py3compat import annotate
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2)

fig.add_trace(go.Box(y=data['v1'], name='v1'), row=1, col=1)
fig.add_trace(go.Box(y=data['v2'], name='v2'), row=1, col=2)
```

```python
#import t-test function
from scipy.stats import ttest_1samp
# calculate p-value
test_stat, p_value= ttest_1samp(data['v1'], popmean= -0.023456, alternative= 'greater')
print('The p-value is', p_value)
```

```python
#import t-test function
from scipy.stats import ttest_1samp
# calculate p-value
test_stat, p_value= ttest_1samp(data['v2'], popmean= -0.023456, alternative= 'greater')
print('The p-value is', p_value)
```

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Two-sample Z-test
z_stat, p_val = ztest(data['v1'], data['v2'], alternative='two-sided')

# Display results
print(f"Z-statistic: {z_stat:.5f}")
print(f"P-value: {p_val:.5f}")

if p_val < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
